```python
commonColumnsAll = ['Datetime', 'username', 'Retweets', 'LikeCount', 'Tweet']
commonColumnsDailygpt1 = ['Tweet_ID', 'Language', 'ReplyCount']
commonColumnsGPTgpt1 = ['URL']
commonColumnsDailyGPT = ['User_Followers']
completlyUseless = ['User_created', 'Verified', 'UserFriends', 'Location', 'Outlinks']
```

```python
# df = GPT
# df2 = gpt1
# df3 = dailygpt
def getAll(df, df2, df3):
    dfDt = df[['User', 'Date', 'Tweet', 'Retweets', 'Likes']]
    dfDt.rename(columns={'User':'user', 'Date':'date', 'Tweet':'tweet', 'Retweets':'retweets', 'Likes':'likes'}, inplace = True) 
    df2Dt = df2[['Username', 'Datetime', 'Text', 'RetweetCount', 'LikeCount']]
    df2Dt.rename(columns = {'Username':'user', 'Datetime':'date', 'Text':'tweet', 'RetweetCount':'retweets', 'LikeCount':'likes'}, inplace = True)
    df3Dt = df3[['user_username', 'tweet_created', 'text', 'retweet_count','like_count']]
    df3Dt.rename(columns = {'user_username':'user', 'tweet_created':'date', 'text':'tweet', 'retweet_count':'retweets', 'like_count':'likes'}, inplace = True)
    newDf = pd.concat([dfDt, df2Dt, df3Dt], ignore_index=True)
    return newDf
```

```python
newDf = getAll(df, df2, df3)
```

```python
len(newDf)
```

```python
def cleanDf(df : pd.DataFrame):
    df.drop_duplicates(subset=['tweet'], inplace=True)
    return df
```

```python
newDf = cleanDf(newDf)
```

```python
newDf = newDf.dropna()
```
